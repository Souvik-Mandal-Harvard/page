## 1. What is Matplotlib?

 [Matplotlib](https://matplotlib.org/) is a library for creating static, animated, and interactive visualizations in Python. There are many librarys out there (e.g. Plotly, Seaborn, Bokeh), and each have their own pros and cons. However, Matplotlib is the most general plotting library that is both flexible and easy to implement. Usually people start by plotting their data with Matplotlib to get a better sense of the data they are working with. Here, we will be using images and videos as our data to visualize how they are formatted.

#### 1.1 Import Matplotlib
Matplotlib is one of many packages you downloaded with Anaconda, so you do not have to separately install it from online. Similar to how we previously imported ```numpy``` you can import ```matplotlib```the same way.

````python
import matplotlib
import matplotlib.pyplot as plt
````

#### 1.1 Create your first plot
To create a line plot, you can use ```plt.plot``` function. As we discussed in our previous project, Python is an object oriented language (i.e. everything is an object). So here, we imported an object called plt, which contains a function called plot. Plot takes multiple arguments, but it requires at least a data as the first argument (or first two arguments). You may be wondering how i know this? You can find out all the information for the librarys simply by searching through their documentations. I would highly recommend you read through matplotlib's [documentation](https://matplotlib.org/contents.html) and [user guide](https://matplotlib.org/tutorials/index.html) so that this library becomes almost second nature. You can also learn more about a library by using the ```? {library}``` command as shown below.

## 2. Import images

Now that you know how to visualize data, let's import image data and see if we can visualize the images as well. We have provided you with one image of cells to look over together.

#### 2.1 Image to matrix
To import the image, we will use matplotlib again, but this time the image object. The image (i.e. ```mpimg```) contains many functions, one of which is imread. This function read images by converting it into a matrix.

````python
import matplotlib.image as mpimg 
# ? mpimg
? mpimg.imread
````

````python
data_path = "img/cells.png"
img = mpimg.imread(data_path)
print(f"Image Shape: {img.shape}")
````

````python
print(img)
````

#### 2.1 Plot the image matrix
We can already see two things from the matrix.

1. The image size is 1182 x 1182 x 3. The first two numbers are you width and height of the window. The third number corresponds to the color channel (i.e RGB - Red, Green, Blue)
2. Each entry is a float that ranges from a value of [0,1]. The higher the number, the more color of that channel is represented for a given pixel. The sum of all three colors from each RGB channel is your final pixel color

<img src="img/img_matrix.png" width="400">

All in all, an image is just a 3-D matrix, with each pixel getting 3 numbers for each of the RGB channel. Sometimes, the number may be an integer rather than a float. In which case, each element will range from [0,255]. Below, I will show you that both [0,1] float and [0,255] int ranges are the same when we plot the image data via matplotlib

````python
print(f"Range [{np.min(img)}, {np.max(img)}]")
plt.imshow(img)
plt.show()
````

````python
img_255 = img*255
img_255 = img_255.astype(int)
print(f"Range [{np.min(img_255)}, {np.max(img_255)}]")
plt.imshow(img_255)
````

## 3. Analyze Images

Now that you know what an image data looks like and how to visualize them using matplotlib, we will try and modify the data. This will give you a better sense of the image data and allow you to analyze things in a new way.

#### 3.1 I only want red cells...
As you can see from the image, we are working with an fluorescent cells marked with either red, green, and blue (how convenient). There may be a time where you are interested in only looking at a specific type of cells, for example the red cells. Well given your understanding of the the image data, how do you think we can do this? Well you know that each pixel has three color channels (Red, Green, and Blue). So if you want to see cells with only red, then you can modify all the pixels with green and blue value to zero. There is multiple ways of doing this, but below is one of the easiest/fastest way of doing it by utilizing the power of numpy arrays.

````python
img_read = np.copy(img) # creates a copy of img data to a new variable
img_read[:,:,1:] = 0
plt.imshow(img_read)
````

## 3. Import Videos

Now that you have mastered images, let's look at videos. You probably already know by now that videos are just series of images, and now know that images are just 3D matrix. Given this, what do you think the video data will look like?

#### 3.1 Importing video data with skvideo

Unfortunately, matplotlib does not have a function to read images. Therefore, we will use a separate library called skvideo to read our video into matrix

````python
import skvideo.io  
````

````python
video_path = "video/CellDivision.mp4"
video = skvideo.io.vread(video_path)  
print(video.shape)
````

#### 3.2 Visualizing one frame of the video

As you can see, a video is a 4D matrix. The first dimension is the frames, and the last three dimensions are similar to an image (i.e. width, heigh, RGB_channel). Given this, we can easily look at the image for specific frame by simply indexing into the 4D matrix as shown below.

````python
fr = 100
plt.imshow(video[fr,:,:,:])
plt.show()
````

## 4. Analyze Videos

As you can see from the image of the video, we have a cell undergoing mitosis. The red marker show where the DNAs are located, and the green marker shows the overall cell structure.

#### 4.1 Where are the DNAs?

Similar to the image, let's try and filter out all the green colors in the video so that we only have red markers. 

````python
video_red = np.copy(video)
video_red[:,:,:,1:] = 0
fr = 100
plt.imshow(video_red[fr,:,:,:])
plt.show()
````
